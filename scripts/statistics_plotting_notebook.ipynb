```python
import pandas as pd
import os
from compile_stats import compare_scatter, compare_errbar, compare_hist
df = pd.read_excel('summary.xlsx', index_col=0, comment='#')  
```

```python
# two major functions for plotting: compare_scatter and compare_errbar
# each need:
#   df - Dataframe 
#   xaxis - Label for xaxis data. Must be a column header in ../data/summary.xlsx file generated. 
#   yaxis - Label for xaxis data. Same as ^
#   xerr (only for compare_errbar) - Label for xaxis's error data.
#   yerr (only for compare_errbar) - Label for yaxis's error data.
# optional:
#   show_dslabels - Boolean to label the dataset (like MoS2_P/ds3) next to each point or not. Default:False
#   material - material name like 'mose2', if given will only show data of this material. 
#   orrient - "p" or "ap", if given will only show data of this stacking.
#   caxis - (only for compare_scatter) Label for color data. Must be a column header in ../data/summary.xlsx file generated.
#     resulting plot has scatter points colored based on this parameter.    
#   dflag - data flag, see below, if given will only show data where dataset is flagged like this.
#   fflag - fit flag, see below, if given will only show data where fit is flagged like this.
#   pflag - geometry partition flag, see below, if given will only data flagged like this.
#   uflag - unwrap flag, see below, if given will only data flagged like this.
```

```python
# EXAMPLE plot showing scatterplot comparision of two variables, here heterostrain and twist.
# filtering data to only include P MoS2 with "good" dataset and "good" fit. No filters applied to data quality or anything.
compare_scatter(df, xaxis="AvgHeteroStrain", 
                yaxis="AvgMoireTwist", show_dslabels=True,
                material='mos2', orrient="p", fflag="good", dflag="good")
```

```python
# EXAMPLE plot showing scatterplot comparision of three variables, here AApercent, AAradius and Material, with 
# material (categorical) repsresented by the color.
# filtering data to only include "good" geometric partitions (so reliable radii/sp widths)
compare_scatter(df, xaxis="AvgHeteroStrain", yaxis="AvgMoireTwist", caxis="DataSetPrefix")
compare_scatter(df, xaxis="AvgHeteroStrain", yaxis="AvgMoireTwist", caxis="FitQualityFlag")
compare_scatter(df, xaxis="AvgHeteroStrain", yaxis="AvgMoireTwist", caxis="DataQualityFlag")
compare_scatter(df, xaxis="AvgHeteroStrain", yaxis="AvgMoireTwist", caxis="PartitionQualityFlag")
```

```python
# EXAMPLE plot showing errbar comparision of AAradius and Moire Twist
# without any filtering of data. Not showing the dslabels next to each point.
compare_errbar(df, yaxis="AvgAAradius",   yerr="ErrAAradius", 
                   xaxis="AvgMoireTwist", xerr="ErrMoireTwist", 
                   material="mos2", orrient='p',
                   show_dslabels=False)

compare_scatter(df, yaxis="AvgAAradius",  xaxis="AvgMoireTwist",
                    material="mos2", orrient='p',
                    caxis="DataQualityFlag",
                    show_dslabels=False)

compare_scatter(df, yaxis="AvgAAradius",  xaxis="AvgMoireTwist",
                    material="mos2", orrient='p',
                    caxis="PartitionQualityFlag",
                    show_dslabels=False)

compare_scatter(df, yaxis="AvgAAradius",  xaxis="AvgMoireTwist",
                    material="mos2", orrient='p',
                    caxis="FitQualityFlag",
                    show_dslabels=False)

compare_errbar(df, yaxis="AvgAAradius",   yerr="ErrAAradius", 
                   xaxis="AvgMoireTwist", xerr="ErrMoireTwist", 
                   material="mos2", orrient='p', pflag="good",
                   show_dslabels=False)
```

```python
compare_scatter(df, yaxis="AAPercent",  xaxis="AvgMoireTwist",
                    material="mos2", orrient='p',
                    caxis="DataQualityFlag",
                    show_dslabels=False)

compare_scatter(df, yaxis="AAPercent",  xaxis="AvgMoireTwist",
                    material="mos2", orrient='p',
                    caxis="FitQualityFlag",
                    show_dslabels=False)

compare_errbar(df, yaxis="AAPercent",      
                   xaxis="AvgMoireTwist", xerr="ErrMoireTwist", 
                   material="mos2", orrient='p', dflag="good", fflag="good",
                   show_dslabels=False)
```
